# **LINEAR REGRESSION**


```python
data.index = pd.to_datetime(data.index,format='%Y-%m-%d')
datas = data.sort_index(ascending=True, axis=0)
datas
```

```python
newdata = pd.DataFrame(index=range(0,len(data)),columns=['Date', 'Close'])
```

```python
for i in range(0,len(datas)):
    newdata['Date'][i] = datas.index[i]
    newdata['Close'][i] = datas['Close/Last'][i]
```

```python
import datetime as dt
newdata['Date'] =  pd.to_datetime(newdata.Date,format='%Y-%m-%d')
#newdata['Date']= newdata['Date'].map(dt.datetime.toordinal)
```

```python
newdata
```

```python
split = (int)(0.65*len(data))
split
```

```python
train = newdata[:split]
train['Date']= train['Date'].map(dt.datetime.toordinal)
test = newdata[split:]
test['Date']= test['Date'].map(dt.datetime.toordinal)
```

```python
x_train = train.drop('Close',axis = 1)
y_train = train['Close']
x_test = test.drop('Close',axis = 1)
y_test = test['Close']
```

```python
from sklearn.linear_model import LinearRegression
price_model = LinearRegression()
price_model.fit(x_train,y_train)
```

```python
y_predict = price_model.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test,y_predict))
print(r2_score(y_test,y_predict))
```

```python
rms=np.sqrt(np.mean(np.power((np.array(y_test)-np.array(y_predict)),2)))
rms
```

```python
test['Predictions'] = 0
test['Predictions'] = y_predict

test.index = newdata[split:].index
train.index = newdata[:split].index

#plt.plot(train['Close'])
plt.plot(test[['Close', 'Predictions']])
```

```python
url = 'https://launchpad.net/~mario-mariomedina/+archive/ubuntu/talib/+files'
ext = '0.4.0-oneiric1_amd64.deb -qO'
!wget $url/libta-lib0_$ext libta.deb
!wget $url/ta-lib0-dev_$ext ta.deb
!dpkg -i libta.deb ta.deb
!pip install ta-lib
import talib
```

```python
data['S_10'] = data['Close/Last'].rolling(window=10).mean()
data['Corr'] = data['Close/Last'].rolling(window=10).corr(data['S_10'])
data['RSI'] = talib.RSI(np.array(data['Close/Last']), timeperiod =10)
data['Open-Close'] = data['Open'] - data['Close/Last'].shift(1)
data['Open-Open'] = data['Open'] - data['Open'].shift(1)
data = data.dropna()
X = data.iloc[:,:10]
```

```python
X
```

```python
Y = np.where(data['Close/Last'].shift(-1) > data['Close/Last'],1,-1)
```

```python
split = int(0.8*len(data))
X_train, X_test, Y_train, Y_test = X[:split], X[split:], Y[:split], Y[split:]
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model = log_model.fit(X_train,Y_train)
```

```python
log_model
```

```python
probability = log_model.predict_proba(X_test)
probability
```

```python
Y_predicted = log_model.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(Y_test,Y_predicted))
```

```python
log_model.score(X_test,Y_test)
```
